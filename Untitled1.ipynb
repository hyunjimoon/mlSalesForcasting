#1. load data

```python
#keras
from keras.models import Sequential
from keras.layers import Dense
import numpy
import matplotlib.pyplot as plt
numpy.random.seed(7)
```

이제 데이터를 로드할 수 있다.

```python
dataset=numpy.loadtxt("pima-indians-diabetes.csv", delimiter=",")
X=dataset[:,0:8]
Y=dataset[:,8]

```

#2. define model


레이어의 수와 유형을 결정하는 것이 중요
첫번째 계층에는 12개의 뉴런이 있고, 8개의 입력변수가 필요
두번째 hidden layer에는 8개의 뉴런이 있고,
마지막 output layer에는 클래스를 예측하는 1개의 뉴런이 있다.

```python
model=Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

#3.compile model


효율적인 수치 라이브러리(so-called backend: Theano, Tensorflow)를 이용하여 모덱을 컴파일<br>training a network means finding the best set of weights to make predictions for this problem.

log loss function 을 이용하고, 이진 분류문제는 keras에서 binary_crossentropy르 정의되고, adam 이라는 stochastic optimization을 이용한다.

```python
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
```

#4. fitting the model

일부 데이터(training set)에서 모델을 실행  
fit()함수를 호출, 두 parameter<br>
--epochs: 
The training process will run for a fixed number of iterations through the dataset called epochs, that we must specify using the epochs argument.<br>
--batch_size: 
the number of instances that are evaluated before a weight update in the network is performed

```python
a=[]
a.append(model.fit(X,Y,epochs=100, batch_size=10))
```

#5. evaluate model

```python
scores =model.evaluate(X, Y)
print("\n%s: %.2f%%" %(model.metrics_names[1], scores[1]*100))
```
